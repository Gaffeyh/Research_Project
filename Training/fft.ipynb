```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): 
    return np.sin(2*np.pi*x)
```

```python
x = np.linspace(0, 2*np.pi, 10000)
y = f(x)
```

```python
plt.plot(x, y)
```

```python
u = np.fft.fft(y)
N = len(x)
k = np.concatenate([ np.arange(0, N//2), [0], np.arange(-N//2+1, 0)]) / 16  # wave numbers
```

```python
plt.plot(k,u.real,label='Real part')
plt.plot(k,u.imag,label='Imaginary part')
plt.legend()
#restrict range for better visualization
plt.xlim(-5,5)
```

```python
def g(s):
    return s**2
```

```python
G = g(x)
plt.plot(x, G)
```

```python
v = np.fft.fft(G)
plt.plot(k,v.real,label='Real part of v(s)')
plt.plot(k,v.imag,label='Imaginary part of v(s)')
plt.legend()
##restrict range for better visualization
plt.xlim(-1,1)
```
